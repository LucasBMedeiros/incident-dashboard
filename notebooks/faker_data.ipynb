```python
from faker import Faker
import pandas as pd

fake = Faker()
data = []
for _ in range(1000):
    data.append({
        'id': fake.uuid4(),
        'name': fake.name(),
        'email': fake.email(),
        'transaction_date': fake.date_this_decade(),
        'amount': fake.random_number(digits=5)
    })
df = pd.DataFrame(data)
df.to_csv('fake_data.csv', index=False)
```
